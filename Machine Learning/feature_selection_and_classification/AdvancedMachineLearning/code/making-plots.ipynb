```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Digits

```python
res_digits = pd.read_csv("results2-digits.csv", index_col=0)
```

```python
res_digits
```

```python
plot_df = pd.DataFrame(res_digits.groupby([ "m"])["score"].agg(np.mean)).reset_index()

ax = sns.lineplot(data=plot_df, x= "m", y = "score")
ax.set(xlabel='Number of features - m', ylabel='Average Score')
ax.set_title("Digits data - GradientBoosting results on validation dataset")
ax.set_label

plt.show()
```

```python
plot_df = pd.DataFrame(res_digits.groupby(["lr", "m"])["score"].agg(np.mean)).reset_index()

ax = sns.lineplot(data=plot_df, hue="lr", x= "m", y = "score")
ax.set(xlabel='Number of features - m', ylabel='Average Score')
ax.set_title("Digits data - GradientBoosting results on validation dataset")
ax.set_label

plt.show()
```

```python
plot_df = pd.DataFrame(res_digits.groupby(["max_depth", "m"])["score"].agg(np.mean)).reset_index()

ax = sns.lineplot(data=plot_df, hue="max_depth", x= "m", y = "score")
ax.set(xlabel='Number of features - m', ylabel='Average Score(m)')
ax.set_title("Digits data - GradientBoosting results on validation dataset")
plt.show()
```

```python
plot_df = pd.DataFrame(res_digits.groupby(["n_est", "m"])["score"].agg(np.mean)).reset_index()

ax = sns.lineplot(data=plot_df, hue="n_est", x= "m", y = "score")
ax.set(xlabel='Number of features - m', ylabel='Average Score')
ax.set_title("Digits data - GradientBoosting results on validation dataset")

plt.show()
```
