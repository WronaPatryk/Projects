```python
from river import datasets
import pandas as pd
import random
import string
import os
import time
```

```python
generator = datasets.synth.Hyperplane()
```

```python


for xi, yi in generator.take(10):
    print(xi)
    print(yi)
```

```python
def generateDataStream(generator, n_in_one_file, n_files, gen_freq_s, LOCAL_PATH, HDFS_PATH):
    
    file_count = 0
    while(file_count < n_files):
        dic = {}
        counter = 0
        for xi, yi in generator.take(n_in_one_file):
            xi['Class'] = yi
            dic[counter] = xi
            counter+=1

        FILE_NAME = ''.join(random.choices(string.ascii_lowercase + string.digits, k=16))
        file_full_path = LOCAL_PATH + "/" + FILE_NAME + ".csv"
        pd.DataFrame.from_dict(dic).transpose().to_csv(file_full_path, header=True, index = None)
        # HDFS
        os.system("hdfs dfs -moveFromLocal " + file_full_path + " " + HDFS_PATH)
        
        file_count += 1
        # WAIT gen_freq_s [s] or n_files
        time.sleep(gen_freq_s)
    
    
    
    
```

```python
# column names
cols = ["V" + str(i) for i in range(len(dic[0]) - 1)]
cols.append("Class")

data = pd.DataFrame.from_dict(dic).transpose()
data.columns = cols

data
```
