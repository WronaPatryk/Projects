```python
import pandas as pd
import string
import random
import numpy as np
from random import sample

def createPrimaryKey(x):
    return ''.join(random.choices(string.ascii_lowercase + string.digits, k=16))

def delayData(data, n_of_delayed):
    data["primary_key"] = 0
    random.seed(44)
    data.primary_key = data.primary_key.map(lambda x: createPrimaryKey(x))
    
    aux = sample(range(0,data.shape[0]), n_of_delayed)
    additional_data = data.iloc[aux, :]

    additional_data.loc[aux, data.columns[:-2]] = None
    data = data.append(additional_data, ignore_index=True)
    data.loc[aux, data.columns[-2]] = None
    return data
    
    
    
    
    
    
    
```
